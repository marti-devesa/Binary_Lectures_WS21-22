Colors:

## 0.2 Git and web-based repositories

### What is git?

Git is nothing else than a version control and management tool. It is a free, open-source tool.

The idea is both end with the practice of having a version of your code called "v32_final_for-reals_I-promise-it-is-the-last_v3" and favour collaboration between developers with a decentralized system. 

GitLab is a web-based repository service. Thus it is designed for tracking, sharing, maintain a documentation, ... It is similar to GitHub, and UIBK provides it as a service for students and staff (you might need to activate your account!).

During this course, we will only use it as a repository to access the notebooks. However it is a very handy tool, used in  almost every programming company or research group nowadays.

### Installing git

$ sudo apt-get install git

Also, the first time git is used it should be configured to your user

$ git config --global user.name "FIRST_NAME LAST_NAME"

$ git config --global user.email "MY_NAME@example.com"
