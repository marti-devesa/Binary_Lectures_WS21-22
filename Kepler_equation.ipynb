## 2 The Kepler Equation

As we have seen in the main lecture, finding a solution for the time dependence of the two-body system is not trivial. In order to overcome such issue, we will employ the Kepler Equation (for a full derivation check Hilditch 2001!).


Image credit: Hilditch 2001

The concept behind everything is that we will start from the second Kepler Law ($r^2\dot{\theta} = L$) and project the ellipse from a relative orbit onto a circle. The only parameter that should matter for you is $E$, the so-called $\textit{eccentric anomaly}$. This allows us to write the conic equation as

<center>$r = a(1-e\cdot cosE)$

and using the second Kepler law ...

<center>$(1-e\cdot cosE) dE =\frac{L}{a\cdot b}dt$

Which is immediately integrated with to

<center>$1-e\cdot sin E=\frac{L}{a\cdot b}t + Constant$

Rearranging using the relations between $L$, the equivalent central mass $M$, the period $P$, the semi-minor ($b$), and semi-major axis ($a$), we get

<center>$E-e\cdot sin E =\frac{2\pi}{P} (t- T_0)$

where we reconverted our constant of integration into the periastron passage time $T_0$. This is a simple, nice equation which relates time with angular position, although in the form of an eccentric anomaly (i.e. Kepler's equation).

You can immediately see, however, that isolating $E$ is not as straightforward. In order to solve the equation, we will need to apply our old friend the Newton-Raphson method, this is

<center> $x_n = x_{n-1} -\frac{f(x_{n-1})}{f'(x_{n-1})}$
