## 5. Eclipsing Binaries 1: Finding the inclination

Once you find a period (check the Lomb-Scargle or PDM tutorials), you can use eclipses in a phaseogram to constrain the inclination of a binary system :D  How does this work?

Geometrically, the eclipse occurs because a disc of a star is covered by the disc of the companion.
